```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
dados = df.sample(500)

dados = dados.reset_index(drop=True)

dados
```

```python
dados.rename(columns = {"PassengerId":"ID_passageiro", "Survived":"Sobreviveu", "Name":"Nome",\
                        "Sex":"Sexo", "Age":"Idade", "Ticket":"Ticket", "Fare":"Passagem", "Cabin":"Cabine",\
                        "Embarked":"Embarcou"}, inplace=True)
```

```python
#Pense que estamos construindo um modelo de ML e queremos prever quem irá morrer e quem irá sobreviver.
#Nesse DF essa informação está na coluna "Sobreviveu", onde quem sobreviveu é 1 e quem morreu é 0.
#Temos interesse em ver como está a distribuição da nosssa variável target(como chamamos em ML), pois se tivermos
#dadis muito desbalanceados, temos que usar algumas tecnicas especificas para balancear os dados.

dados.Sobreviveu.value_counts()
```

```python
#poderiamos ver em percentual:

dados.Sobreviveu.value_counts()/dados.shape[0] #numero de pessoas a bordo
```

```python
#agora vamos selecionar a coluna sexo e verificar quantos são homens

morreram[morreram.Sexo=='male'].shape
```

```python
#Viram que na variavel cabine tem valores nan (Not a number)? E em outras variáveis também tem?
#No dia-a-dia do cientista é muito importante conhecer as variáveis, saber o que podem ser esses valores NaN, saber
#o % de missing em uma variável é também saber como tratá-los:

dados.isnull()
```

```python
#Para o nosso estudo vamos deletar as observações(linhas) que possuem valores faltantes:

print("Qtd nulos antes do drop:\n", dados.isnull().sum())
dados = dados.dropna()
print('----------------------------------')
print("Qtd nulos após o drop:\n", dados.isnull().sum())
```

```python
Função Map
```

```python
# Para construir modelos de ML transfromamos todos os campos que são strings em campos numéricos. Vejam o que 
#podemos fazer, por exemplo, pra variável sexo.

print(dados.Sexo[:5])

dados.sexo = dados.Sexo.map({"male":0,"female":1})
```

```python
#Podemos fazer agrupamentos aqui, assim como fizemos em SQL.Lembram da função group by lá?
dados.groupby(by = "Sexo")
```

```python
#Esse groupby faz algo desse tipo aqui:
dados[dados.sexo==1][:5]
```

```python
dados[dados.sexo==0][:5]
```

```python
#Mesmo salvando em uma variavel , não conseguimos ver o agrupamento feito pelo python:

dff = dados.groupby(by = "Sexo")
dff
```

```python
#Uma exporessão lambda permite escrever funções anonimas, em tempo de codigo, usando apenas uma linha de codigo.
#Onde x1, x2... são variáveis que representem o argumento da função e expr é qualquer expressão válida em Python,
#envolvendo essas variáveis -- lambda x,y: np.sum(x,y)

dados.groupby(by = "Sexo").apply(lambda x: x.Idade.mean() )
```

```python
dados.groupby(by = "Sexo").apply(lambda x: x.Sobreviveu.value_counts() )
```

```python
#Poderiamos também aplicar uma função nossa, customizada:
def f(x):
    return np.max(x) - np.min(x)

dados.groupby(by = "Sexo").apply(lambda x: f(x.Idade) )
```

```python
#Podemos usar o apply sem usar o groupby por exemplo, apenas aplicar nas series diretamente:

dados[["Passagem","Idade"]].apply(lambda s: s.mean())
```

```python
#Outra função muito usada também, além de apply é a aggragate. Aqui usamos a agg para aplicar a função min() em 
#todas as colunas:
dados.groupby("Cabine").agg('min')
```

```python
#Podemos usar a agg para aplicar uma função especifica para cada coluna de um data frame

def moda(s):
    ss = s.value_counts() 
    return ss.idxmax()

dados.groupby('Cabine').agg({"Sobreviveu": 'sum', "Idade":['min', 'max'], "Embarcou":moda})
```

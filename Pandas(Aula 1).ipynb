DataFrame e Series

Podemos pensar nos DataFrames como um arquivo  csv, planilha de Excel, ou até mesmo uma tabela de um banco de dados, como vimos no módulo passado. Essa estrutura de dados bidimensional é composta por três elementos principais: linhas, colunas e índices.

Cada uma das colunas do DataFrame é uma Serie. As series podem ser comparadas as listas, que vimos nos primeiros módulos.

Serie é um array unidimensional que pode armazenar qualquer tipo de dados e possui índices.

Exemplo:
a 10

b 20

c 30

d 40

e 50

f 60

dtype: int64

Tanto as Series quanto os DataFrames possuem funções especificas e nós veremos boa parte delas hoje!

```python
import pandas as pd
import numpy as np
```

```python
Para o nosso estudo de pandas usaremos um conjunto de dados do titanic. 
Esse arquivo deve ser baixado e por simplicidade, colocado na mesma pasta onde você roda seu notebook.
```

```python
#Pandas possui várias funções para ler dados. Tente ai na sua casa ler um arquivo excel por exemplo
df = pd.read_csv("titanic.csv")  # sep = 'separador' - "," , ";"

#a função head, nos mostra as 5 primeiras linhas do dataframe:
df.head()
```

```python
#Podemos customizar essa função passando como parametro a quantidade de linhas que queremos ver:
df.head(20)
```

```python
#Ou ainda poderiamos ter feito um slicing no df:

df[:10]
```

```python
#Analisando dimensões do nosso dataframe:

df.shape

#891 linhas, 12 colunas
```

```python
#Função Sample
#Supondo que nosso conjunto de dados fosse muito grande, poderiamos trabalhar com uma amostra
#nesse caso é importante selecionarmos uma amostra aleatoria para termos representatividade dos dados:

dados = df.sample(500)

dados = dados.reset_index(drop=True)

dados
```

```python
#Converte para um array bidimensional
dados.values
```

```python
#Retorna todos os nomes das colunas:
dados.columns
```

```python
#Função DROP
#Não utilizaremos as colunas Pclass, SibSp e Parch, pois não nos trazem muita informação, então vamos descartá-las:

dados = dados.drop(["Pclass", "SibSp", "Parch"], axis=1)
dados.head()
```

```python
#Não é uma boa prática temos colunas cujos nomes estão separados por espaço. É melhor separarmos por "_".
#Ainda, como os nomes estão em inglês vamos renomear para o português para ficar mais fácil o entendimento:

dados.rename(columns = {"PassengerID":"ID_passageiro", "Survived":"Sobreviveu", "Name":"Nome",\
                        "Sex":"Sexo", "Age":"Idade", "Ticket":"Ticket", "Fare":"Passagem", "Cabin":"Cabine",\
                        "Embarked":"Embarcou"}, inplace=True)
```

```python
dados.columns
```

```python
#Acessando os valores de uma coluna/serie: (df[nomecoluna],df.nome_coluna)

print(dados["Passagem"])

print('\n*******************\n')

print('Tipo de estrutura onde guardamos os valores da coluna: ',type(dados.Passagem))
```

```python
#Conseguimos acessar elementos especificos das colunas através do índice:

print(dados.Passagem)
print('\n*****************\n')
print(dados.Passagem[4])
```

```python
#Podemos ordenar os valores das colunas tanto por ordem decrescente, como por ordem crescente:

print("Ordem crescente\n\n", dados.Passagem.sort_values(ascending=True))
print('\n*********************\n')
print("Ordem decrescente\n\n", dados.Passagem.sort_values(ascending=False))
```

```python
#Podemos fazer também scling nas colunas, assim como fizemos com as listas e arrays:

dados.Passagem[5:10]
```

```python
dados.head
```

```python
#Analisando os tipos de dados de cada coluna do DF
dados.dtypes
```

```python
#Nós não usaremos a coluna ID_passageiro, pois ela não trás informação nenhuma. Geralmente quando criamos modelos,
#tiramos essas colunas de identificação(Id's)

#Mas antes de deletarmos essa coluna, vamos ver como alterar o tipo de dados de uma coluna:

dados.PassengerId = dados.PassengerId.astype(float)

dados.head()
```
